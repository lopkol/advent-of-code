## Problem 1 + 2

```python
next_chars = {
    'm': ['u'],
    'u': ['l'],
    'l': ['('],
    'd': ['o'],
    'o': ['n', '('],
    'n': ["'"],
    "'": ['t'],
    't': ['(']
}

def calculate(mult_str):
    numbers = mult_str[4:-1].split(',')
    return int(numbers[0]) * int(numbers[1])

def evaluate_condition(cond_str):
    if cond_str == 'do()':
        return True
    if cond_str == "don't()":
        return False
    raise Exception(f'Invalid conditional string: {cond_str}')
```

```python
# solves both problems, depending on the second param
def solve2(input_filename, conditionals_enabled):
    f = open(f'../inputs/{input_filename}.txt', 'r')

    res = 0
    enable_mult = True
    stm, num1_chars, num2_chars = '', 0, 0
    
    while True:
        char = f.read(1)
        if not char:
            break
        if char == 'm':
            # we start a new multiplication str
            stm, num1_chars, num2_chars = 'm', 0, 0
            continue
        if char == 'd':
            # we start a new conditional str
            stm, num1_chars, num2_chars = 'd', 0, 0
            continue

        # There is no switch :'( 
        if len(stm) == 0:
            stm, num1_chars, num2_chars = '', 0, 0
        elif stm[-1] in next_chars.keys():
            if char not in next_chars[stm[-1]]:
                stm, num1_chars, num2_chars = '', 0, 0
            else:
                stm += char
        elif stm[-1] == '(':
            if stm[0] == 'm' and char in numbers:
                num1_chars = 1
                stm += char
            elif stm[0] == 'd' and char == ')':
                stm += char
                enable_mult = evaluate_condition(stm)
                stm, num1_chars, num2_chars = '', 0, 0
            else:
                stm, num1_chars, num2_chars = '', 0, 0
        elif stm[-1] in numbers and num1_chars > 0:
            if char in numbers and num1_chars < 3:
                stm += char
                num1_chars += 1
            elif char == ',':
                stm += char
                num1_chars = 0
            else:
                stm, num1_chars, num2_chars = '', 0, 0
        elif stm[-1] == ',':
            if not char in numbers:
                stm, num1_chars, num2_chars = '', 0, 0
            else:
                num2_chars = 1
                stm += char
        elif stm[-1] in numbers and num2_chars > 0:
            if char in numbers and num2_chars < 3:
                stm += char
                num2_chars += 1
            elif char == ')':
                stm += char
                if enable_mult or not conditionals_enabled:
                    res += calculate(stm)
                stm, num1_chars, num2_chars = '', 0, 0
            else:
                stm, num1_chars, num2_chars = '', 0, 0
                
    f.close()
    return res
```
